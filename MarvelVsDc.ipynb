```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
df = pd.read_csv('db.csv', encoding='latin1')
```

```
print(df.columns)
```

```
print(df.head(3))
sns.set_style('dark')
```

```
df.head(3)
```

```
sns.countplot(x = df['Company'])
plt.title('Number of Movies')
plt.ylabel('Count')
```

```
df_2 = df.sort_values('Gross Worldwide', ascending=False)
sns.barplot(x = df_2['Original Title'], y = df_2['Gross Worldwide'], data=df_2, order = df_2['Original Title'][:10] )
plt.xticks(rotation = 90)
plt.title('Top 10 Highest Grossing')
plt.xlabel('Movie Title')
plt.ylabel('Worldwide Gross')
```

```
my_pal = {"Marvel": "Red", "DC": "#add8e6"}
sns.boxplot(x = df_2['Company'], y = df_2['Gross Worldwide'], data=df_2, hue = 'Company', palette = my_pal)
plt.xticks(rotation = 90)
plt.title('Gross for all Movies')
plt.xlabel('Company')
plt.ylabel('Worldwide Gross')
```

```
df_3 = df.sort_values('Rate', ascending=False)
sns.barplot(x = df_3['Original Title'], y = df_3['Rate'], data=df_3, order = df_3['Original Title'][:10], palette = 'Blues_d')
plt.xticks(rotation = 90)
plt.title('Ranked by IMDB')
```

```
df_4 = df.sort_values('Metascore', ascending=False)
sns.barplot(x = df_4['Original Title'], y = df_4['Metascore'], data=df_4, order = df_4['Original Title'][:10] , palette = 'Greens_d')
plt.xticks(rotation = 90)
plt.title('Metacritic Score')
```

```
df = df.astype({'Budget': 'int64'})
df['Profit'] = df['Gross Worldwide'] - df['Budget']
df.columns
```

```
df_5 = df.sort_values('Profit', ascending=False)
sns.barplot(x = df_5['Original Title'], y = df_5['Profit'], data=df_5, order = df_5['Original Title'][:10] , palette = 'Reds_d')
plt.xticks(rotation = 90)
plt.title('Profit Created')
```
